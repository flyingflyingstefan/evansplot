```python

# You will need to configure this to point to the "examples" directory where the data files are found
EXAMPLES_DIR = "jason/evansplot/examples"

gcm_ds = xr.open_dataset(os.path.join(EXAMPLES_DIR,"gcm_co2_pre.nc"))
```

```python

p = evansplot.evans_plot(gcm_ds, 
                     title = "Change in Precipitation",
                     hue_var="anpre48_00", 
                     sat_var="per_anpre48_00",
                     num_sat_levels=4,
                     min_sat=-20,max_sat=20,
                     hue_label="mm", sat_label="%",
                     )
```
