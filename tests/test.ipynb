```python
# generate some test data
lat_steps = 36 * 4
lat_au = np.linspace(-45, -10, lat_steps)
lon_steps = (154 - 113 + 1 ) * 4
lon_au = np.linspace(113, 154, lon_steps)
lat_gradient = np.linspace(0, 1, len(lat_au))
lat_gradient = np.expand_dims(lat_gradient, axis=1)
lat_gradient = np.repeat(lat_gradient, len(lon_au), axis=1)
lon_stripe = lon_au.copy()
lon_stripe = np.expand_dims(lon_stripe, axis=0)
lon_stripe = np.repeat(lon_stripe, len(lat_au), axis=0)
ds = xr.Dataset(
    {
        "lat_gradient": (["lat", "lon"], lat_gradient),
        "lon_stripe": (["lat", "lon"], lon_stripe),
    },
    coords={
        "lat": lat_au,
        "lon": lon_au,
    },
)
```

```python

fig = evansplot.evans_plot(ds, "lon_stripe", "lat_gradient", 
                     title="Test Pattern Aus", 
                     hue_threshhold_step_size=10, 
                     num_sat_levels=4, 
                     figsize=(8, 10))
```

```python

fig = evansplot.evans_plot(ds, "lon_stripe", "lat_gradient", 
                     title="Test Pattern Aus Cyclic", 
                     hue_threshhold_step_size=5, 
                     num_sat_levels=3, 
                     cbar="cyclic", 
                     figsize=(10, 7))
```

```python

fig = evansplot.evans_plot(ds, "lon_stripe", "lat_gradient", 
                     title="Test Pattern Aus Vertical", 
                     hue_threshhold_step_size=10, 
                     num_sat_levels=4, 
                     cbar="vertical", 
                     figsize=(8, 7))
```
